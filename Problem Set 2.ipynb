### 1. Population Growth and Dynamic Inefficiency in the OLG Model.

#### a) Derive an equation for the evolution of capital per worker in this economy between periods t and t + 1.

Assume population dynamics are given by $L_{t+1} = \Xi_{t+1}L_t$ where $L$ is the population of young households and $\Xi\equiv(1+\xi)$ is the growth factor for the population of young people. Additionally, assume logarithmic utility and a Cobb-Douglas aggregate production function.

$$f(k_t) \equiv F(K_t,L_t)/L_t = F(K_t/L_t,1)$$
$$W_t = f(k_t) - k_t f'(k_t)$$
$$r_t = f'(k_t)$$

$$ \text{Let  } F(K_t,L_t) = K^\alpha L^{1-\alpha} \Rightarrow f(k_t) = \Big(\frac{K}{L}\Big)^\alpha = k^\alpha $$

$$ \text{So, } W_t = k_t^\alpha - k_t \alpha k_t^{\alpha - 1} = (1 - \alpha)k_t^\alpha $$ $$ r_t = \alpha k_t^{\alpha - 1} $$

$$ u'(c_{1,t}) = \beta R_{t+1}u'(c_{2,t+1}) $$

$$ c_{1,t} = \frac{W_{1,t}+(W_{2,t+1} / R_{t+1})}{1+\beta} $$ $$W_{2,t+1} = 0 \text{  since the old generation does not work and make a wage}$$

$$ c_{1,t} = \frac{W_{1,t}}{1+\beta} $$

$$ \text{Let  } a_{1,t} \text{  be the assets per young household after consumption in period 1} $$

$$ a_{1,t} = W_{1,t}-c_{1,t} $$
$$ a_{1,t} = W_{1,t}-\frac{W_{1,t}}{1+\beta} $$
$$ a_{1,t} = W_{1,t}\Big(\frac{\beta}{1+\beta}\Big) $$
$$ a_{1,t} = (1-\alpha)k_t^\alpha \Big(\frac{\beta}{1+\beta}\Big) $$

$$ k_{t+1} = \frac{a_{1,t}}{\Xi_t} $$

$$ k_{t+1} = k_t^\alpha \Big[ \frac{(1-\alpha)\beta}{\Xi_t(1+\beta)}\Big] $$

$$ \frac{dk_{t+1}}{dk_t} = k_t^{\alpha - 1} \Big[\frac{\alpha(1-\alpha)\beta}{\Xi_t(1+\beta)}\Big] $$

### 2. Social Security and the Baby Boom.

Let there be a "baby boom" so that $L_t=1$ and $L_{t+s}=P_s \ \forall \ s \in \mathbb{Z} \ , s\neq0 \text{  where  } P_s<<1$. In other words, the generation born at time t is larger than all other generations.

By keeping per capita taxes constant,  yield the following relationship:

$$ z_{2,t+1}=Z_{2,t+1} \ / \ L_t$$
$$ z_{2,t+1}=-Z_{1,t+1} \ / \ L_t$$
$$ z_{2,t+1}=\Big( \frac{-Z_{1,t+1}}{L_{t+1}} \Big) \Big( \frac{L_{t+1}}{L_t} \Big) $$
$$ z_{2,t+1}=-z_{1,t+1}(P_1) $$

Here we can see that the older generation will only receive a portion of the per capita benefits paid by the younger generation, as $P_1 < 1$.

By keeping per capita benefits constant, yield the following relationship:

$$ \bar{Z}_t = Z_{1,t} + \frac{Z_{2,t+1}}{R_{t+1}} $$
$$ \bar{Z}_t = Z_{1,t}(1 - 1 \ / \ R_{t+1}) $$
$$ \bar{Z}_t = \frac{r_{t+1}Z_{1,t}}{R_{t+1}} $$

With the given population assumptions, we can compare $\bar{Z}_t$ at time $t$ and time $t+s$

$$ \frac{\bar{Z}_t}{L_t} = \frac{r_{t+1}Z_{1,t}}{R_{t+1}L_t} $$

$$ \bar{z}_t = \frac{r_{t+1}Z_{1,t}}{R_{t+1}} $$

$$ \frac{\bar{Z}_{t+s}}{L_{t+s}} = \frac{r_{t+1+s}Z_{1,{t+s}}}{R_{t+1+s}L_{t+s}} $$

$$ \bar{z}_{t+s} = \frac{r_{t+1+s}Z_{1,t+s}}{R_{t+1+s}} \frac{1}{P_s} $$

Since $P_s < 1$, then all else constant $\bar{z}_{t+s} > \bar{z}_t$.
